```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('model_errors.csv')
df = df.groupby(['wrong_errortype', 'correct_errortype']).size()
model_errors_count = df.to_frame(name = 'count').reset_index()
model_errors_count = model_errors_count[model_errors_count['count'] > 10]
model_errors_count = model_errors_count.groupby(['wrong_errortype', 'correct_errortype']).sum().unstack(fill_value=0)

model_errors_count.to_csv('error_prediction_model_errors.csv')
model_errors_count
```
